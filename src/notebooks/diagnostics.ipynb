```python
import pandas as pd

from pathlib import Path

df_path = Path.cwd().parent.parent / "data" / "daily_returns_labeled.csv"

print(df_path)
df = pd.read_csv(df_path)

```

```python
for regime in df["Regime"].unique():
    subset = df[df["Regime"] == regime]
    plt.hist(subset["SP500"], bins=20, alpha=0.5, label=f"{regime} SP500")
    plt.hist(subset["10Y"], bins=20, alpha=0.5, label=f"{regime} 10Y")
```

```python
plt.legend()
plt.title("Monthly Return Distributions by Regime")
png_path = Path.cwd().parent.parent / "output" / "monthly_return_distributions_by_regime.png"
plt.savefig(png_path)

```
